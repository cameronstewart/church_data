```
import feedparser
import pandas as pd
```

# Source Data

```
#RSS EXAMPLE
rawrss = [
    'http://balgowlahanglican.com.au/feed/podcast/',
    'https://inchurch.com.au/podcast?format=rss',
   # 'http://www.sovgrace.org.au/podcasts/sgcollege/rss.xml',
    #http://www.allsouls.net.au/itunes.rss'
    
    
    ]

feeds = [] # list of feed objects
for url in rawrss:
    feeds.append(feedparser.parse(url))

posts = [] # list of posts [(title1, link1, summary1), (title2, link2, summary2) ... ]
for feed in feeds:
    for post in feed.entries:
        posts.append((post.title, post.link, post.summary, post.guid, post.link))

df = pd.DataFrame(posts, columns=['title', 'link', 'summary', 'guid', 'link']) # pass data to init
```

```
#ITUNES RSS EXAMPLE

#RSS EXAMPLE
rawrss = [
    'http://www.allsouls.net.au/itunes.rss',
    'https://www.cch.net.au/podcast.xml'
    
    
    ]

feeds = [] # list of feed objects
for url in rawrss:
    feeds.append(feedparser.parse(url))

posts = [] # list of posts [(title1, link1, summary1), (title2, link2, summary2) ... ]
for feed in feeds:
    for post in feed.entries:
        posts.append((post.title, post.author, post.subtitle, post.summary))

df_itunes = pd.DataFrame(posts, columns=['title', 'author', 'subtitle', 'summary']) # pass data to init
```
